```python
import bokeh
import pandas as pd
import numpy as np
from bokeh.charts import Bar, Histogram
from bokeh.io import show, output_notebook

output_notebook()
csv = pd.read_csv("media/documents/testdocs/FL_insurance_sample.csv")
df = pd.DataFrame(csv)
_, count = np.unique(np.array(csv['county']), return_counts=True)
p = Bar(df, values='county', agg='count', width=400)
show(p)
```
